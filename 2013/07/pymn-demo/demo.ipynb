## Conway's Game of Life

Matplotlib provides matlab style plotting. From Travis Oliphant

```python
size = 30
GRID = (np.random.rand(size, size) > 0.5).astype(np.uint8)
indx = np.r_[0:size]
up = np.roll(indx, -1)
down = np.roll(indx, 1)
figsize(7,7)
```

```python
neighbors = GRID[up,:] + GRID[down,:] + GRID[:,up] + GRID[:,down] + \
            GRID[np.ix_(up, up)] + GRID[np.ix_(down, down)] + \
            GRID[np.ix_(up, down)] + GRID[np.ix_(down, up)]
GRID = (neighbors == 3) | (GRID & (neighbors == 2))

plt.matshow(GRID, cmap="Greys")
plt.show()
```

## Predator Prey Model  
$$ x' = x(\alpha - \beta y) $$
$$ y' = y(\gamma  + \beta \delta x)  $$

A mathematical model of rabbits and foxes.
http://wiki.scipy.org/Cookbook/LoktaVolterraTutorial

```python
from scipy.integrate import odeint

a, b, c, d = 1.0, 0.1, 1.5, 0.75

def dxdy(X, t=0):
    x, y = X
    return [x*(a - b*y), -y*(c - d*b*x)]

rabbits, foxes = odeint(dxdy, [10, 5], np.linspace(0, 20, 1000)).T

plt.plot(rabbits, linewidth=3)
plt.plot(foxes, linewidth=3, color="red")
plt.ylabel("Population", fontsize=24)
plt.xlabel("Time", fontsize=24)
```

## FiPy -- Finite Volume PDE solver

```python
import fipy as fp
nx = 20
ny = nx
dx = 1.
dy = dx
L = dx * nx
mesh = fp.Grid2D(dx=dx, dy=dy, nx=nx, ny=ny)
phi = fp.CellVariable(name = "solution variable",
                   mesh = mesh,
                   value = 0.)
D = 1.
eq = 0 == fp.DiffusionTerm(coeff=D)

X, Y = mesh.faceCenters
facesTopLeft = ((mesh.facesLeft & (Y > L / 2))
             | (mesh.facesTop & (X < L / 2)))
facesBottomRight = ((mesh.facesRight & (Y < L / 2))
                 | (mesh.facesBottom & (X > L / 2)))

phi.constrain(0.0, facesTopLeft)
phi.constrain(1.0, facesBottomRight)

eq.solve(var=phi, dt=1e-5)
plt.matshow(phi.value.reshape(20, 20))
```
